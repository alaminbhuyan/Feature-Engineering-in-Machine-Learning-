```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/Supervised Learning/Some Important Topic for Machine Learning/Features Engineering /Outliers find/heights.csv")
df
```

# Detect outliers using **percentile**

```python
max_thresold = df['height'].quantile(0.95)
max_thresold
```

```python
df[df['height']>max_thresold]
```

```python
min_thresold = df['height'].quantile(0.05)
min_thresold
```

```python
df[df['height']<min_thresold]
```

# **Remove outliers**

```python
df[(df['height']<max_thresold) & (df['height']>min_thresold)]
```

### Explore samples that are above 99.90% percentile and below 1% percentile rank

```python
min_thresold2,max_thresold2 = df2.price_per_sqft.quantile([0.01,0.999])
min_thresold2,max_thresold2
```

```python
# minimum outliers
df2[df2['price_per_sqft']<min_thresold2]
```

```python
# maximum outliers
df2[df2['price_per_sqft']>max_thresold2]
```

# **Remove outliers**

```python
new_df = df2[(df2.price_per_sqft<max_thresold2) & (df2.price_per_sqft>min_thresold2)]
new_df
```
