### Find out particular Node Importance:

#### `Formula:`

    node_Num = importance of particular node / sum(all node importance)
    
### To find out the importance of a Node:

    Ni = (N_t / N) * [impurity - ((N_t_r / N_t) * right_node_impurity) - (N_t_l / N_t) * left_node_impurity)]


where:
    
    1) Ni = Node importance
    
    2) N_t = Number of rows contain that particular node.
    
    3) N = Total rows
    
    4) impurity = "Gini" or "Entropy" value for that node
    
    5) N_t_r = Number of rows of the right node
    
    6) N_t_l = Number of rows of the left node
    
    7) right_node_impurity = "Gini" or "Entropy" value for right node
    
    8) left_node_impurity = "Gini" or "Entropy" value for left node

### Here, number of features is two: 0 and 1. So decision tree split based on X[1] and X[0]

### Calculation features Importance in decision tree and random forest

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
import sklearn
```

```python
print("sklern version: ", sklearn.__version__)

# sklern version:  1.2.0
```

```python
X, y = make_classification(n_samples=5, n_classes=2,
                               n_features=2, n_informative=2, n_redundant=0,
                               random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

## In random forest we just take the average

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=2)
rf.fit(X,y)
```

```python
rf.feature_importances_
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)

print((1 + 0.555)/2)
```
