```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('dataset/train.csv')
```

```python
df.head()
```

```python
# We are not doing project. So, I just take the necessary column to learning purpose

df.drop(columns=['PassengerId','Name','Ticket','Cabin'], inplace=True)
```

```python
df.head()
```

```python
# Check null value

df.isnull().sum()
```

```python
# Step 1 -> train/test/split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']),
                                                    df['Survived'],
                                                    test_size=0.2,
                                                    random_state=2)
```

## Filling Missing value using SimpleImputer

```python
# Applying imputation

si_age = SimpleImputer() # By default fill missing value by the "mean"

si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X=X_train[['Age']])

X_train_embarked = si_embarked.fit_transform(X=X_train[['Embarked']])

X_test_age = si_age.transform(X=X_test[['Age']])

X_test_embarked = si_embarked.transform(X=X_test[['Embarked']])
```

## Apply Encoding technique

```python
# one hot encoding Sex and Embarked

# sparse = False means it will not return any sparse matrix rather than it will return an array
# handle_unknown = "ignore" means if new category come in future it will be ignore

ohe_sex = OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X=X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X=X_train_embarked)

X_test_sex = ohe_sex.transform(X=X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X=X_test_embarked)
```

```python
# I drop those column those are transform till now for training

X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
# I drop those column those are transform till now for testing

X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem, X_train_age, X_train_sex, X_train_embarked), axis=1)
X_test_transformed = np.concatenate((X_test_rem, X_test_age, X_test_sex, X_test_embarked), axis=1)
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()

clf.fit(X=X_train_transformed, y=y_train)
```

```python
y_pred = clf.predict(X=X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_true=y_test, y_pred=y_pred)
```
